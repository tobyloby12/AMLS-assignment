```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image
from keras.preprocessing.image import load_img
```

```python
df_labels = pd.read_csv('./dataset/label.csv')

print(df_labels.head(5))
print(f'Number of images: {df_labels.size}')
print(f"Unique labels: {df_labels['label'].unique()}")
```

```python
# adding column for binary classification so that computer knows if there is a tumor or not
df_labels['tumor_present'] = np.where(df_labels['label'] == 'no_tumor', 'no_tumor', 'tumor')

print(df_labels.head(5))

print(f"\n Number of labels with each tumor: \n  {df_labels['label'].value_counts()} \n tumor: {df_labels['tumor_present'].value_counts()[0]}")
```

```python
df_labels['label'].value_counts().plot.bar()
plt.title('Showing class distribution of data')
plt.xlabel('Class name')
plt.ylabel('Number of occurances')
plt.show()
```

```python
# showing example image and checking for size
img_path = './dataset/image/IMAGE_0000.jpg'
img = Image(img_path)
img = load_img(img_path)
display(img)
print(f"Image size: {img.size}")
```
